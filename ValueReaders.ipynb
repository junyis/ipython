## 주식투자경력 및 원칙
> 주식투자경력
- 소규모로 7년정도 투자 후 (~2017.03~) 이후 중규모 1년 투자 중 (현재 23종목 보유 중)

> 기본투자전략(원칙)
- 종합주가지수에따른 투자 비중조절 전략
- 항상 전체투자가능금액의 10%를 현금으로 보유
- 종합주가지수 전고점 대비 70~80%의 주식비중 유지
- 주식과 10%현금을 제외한 나머지 금액은 채권ETF로 보유
- 종합주가지수 전고점 대비 이상10% 하락시 10% 추가 종목 매수
- 종합주가지수 전고점 대비 이상10% 상승시 10% 비중의 종목 매도 후 채권 구매
- 1회 투자시 1개종목에 전체 투자가능금액의 3% 정도 매수
- 종목 선택시에는 아래 4가지 체크사항 점검 후 매수 <br>
1> "재무제표 모르면 주식투자 절대로 하지마라" 사경인 저 기준 SRIM Value 평가<br>
2> "밀레니얼머니" 패트링오쇼너시 저 기준 4가지 현금흐름 체크리스트<br>
3> "내꿈은 전업투자자" 정민경 저 기준 사업보고서 가격(P) 판매량(Q) 비용(C) 분석<br>
4> "불곰의 주식투자 불패공식" 박선목, 불곰 저 기준 회사의 정성적 분석<br>
